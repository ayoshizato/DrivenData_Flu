```python
# Determining what features will be removed bue to the excessive number of NaN values.
# By unsing a threshold of 15% of NaN, the columns 15, 26, 34, 35 have been dropped out.

list_rem_train = train_feat.isnull().sum() > len(train_feat) * 0.15
list_rem_test = test_feat.isnull().sum() > len(test_feat) * 0.15
print(list_rem_train, list_rem_test )
```

```python
# Dropping out the features with excessive NaN
feat_rem = ['health_insurance','employment_industry','employment_occupation','income_poverty']
train_feat_rem = train_feat.drop(feat_rem, axis = 1)
test_feat_rem = test_feat.drop(feat_rem, axis = 1)
train_feat_rem.columns, train_feat_rem.shape, test_feat_rem.columns, test_feat_rem.shape
```

```python
# Checking the new list of columns and their number of NaN
train_feat_rem.isnull().sum()
```

```python
# Checking the new list of columns and their number of NaN
test_feat_rem.isnull().sum()
```

```python
# Dropping the rows containing NaN
train_feat_rem.dropna(inplace = True)
train_feat_rem.isnull().sum()
```

```python
# Dropping the rows containing NaN
test_feat_rem.dropna(inplace = True)
test_feat_rem.isnull().sum()
```

```python
# Checking if the states in the Train and Test datasets are the same
for ix, column in enumerate (train_feat_rem.iloc[:,1:].columns):
    if sorted(list(test_feat_rem[column].unique())) != sorted(list(train_feat_rem[column].unique())):
        print(ix, column, 'DOES NOT MATCH!')
    else:
        print(ix, column, 'matches')
```

```python
# Encoding the Categories
from sklearn.preprocessing import OrdinalEncoder
train_obj = []
test_obj = []
for ix,column in enumerate(train_feat_rem.columns):
    if train_feat_rem[column].dtypes == object:
        train_obj.append(column)
        new_column = column + '_code'
        ord_enc = OrdinalEncoder()
        train_feat_rem[new_column] = ord_enc.fit_transform(train_feat_rem[[column]])
        
for ix,column in enumerate(test_feat_rem.columns):
    if test_feat_rem[column].dtypes == object:
        test_obj.append(column)
        new_column = column + '_code'
        ord_enc = OrdinalEncoder()
        test_feat_rem[new_column] = ord_enc.fit_transform(test_feat_rem[[column]])
```
